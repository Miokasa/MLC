```python
data = {
    'Department': ['HR', 'Engineering', 'Marketing', 'Finance', 'Sales', 'HR', 'Engineering', 'Marketing', 'Finance', 'Sales'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'Salary': [70000, 80000, 60000, 90000, 75000, 72000, 85000, 62000, 91000, 76000],
    'Years': [5, 6, 3, 8, 7, 4, 5, 2, 9, 6]
}

df = pd.DataFrame(data)

grouped = df.groupby('Department')['Salary'].mean()

grouped


grouped_multi = df.groupby(['Department', 'Years'])['Salary']
grouped_multi.mean()
```

```python
agg_data = df.groupby('Department').agg({
    'Salary': ['mean', 'sum'],
    'Years': ['mean', 'max']
})

agg_data
```

```python
df['Salary_mean'] = df.groupby('Department')['Salary'].transform('mean')

df
```

```python
filtered = df.groupby('Department').filter(lambda x: x['Salary'].mean() > 75000)

filtered
```

```python
# index, columns, values, aggfunc

pivot = df.pivot_table(values = 'Salary', index = 'Department', columns = 'Years', aggfunc = ['mean','sum'], fill_value=0)

pivot
```
